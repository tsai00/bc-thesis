```python
import pandas as pd
```

```python
# Loading data

url = 'https://www.bezrealitky.cz/api/record/markers?&estateType=byt'
data = pd.read_json(url)
data.shape
```

```python
# First 5 rows

data.head()
```

```python
# Basic information about columns

data.info()
```

```python
# Inspecting 'keyAdvertType' column

data['keyAdvertType'].value_counts()   # Only one value -> column can be dropped
```

```python
# Inspecting 'orderPriority' column

data['orderPriority'].describe()    # Value is always 0 -> column can be dropped
```

```python
# Inspecting 'type' column

data['type'].value_counts()    # Source of ad -> column can be dropped
```

```python
# Splitting column 'timeOrder'

timeOrderSplitted = pd.json_normalize(data['timeOrder'])
```

```python
# Joining dataframes

data = data.join(timeOrderSplitted)
data.head()
```

```python
# Splitting column 'advertEstateOffer'

data['lat'] = data['advertEstateOffer'].apply(lambda x: eval(x[0]['gps'])['lat'])
data['lng'] = data['advertEstateOffer'].apply(lambda x: eval(x[0]['gps'])['lng'])

data['price'] = data['advertEstateOffer'].apply(lambda x: x[0]['price'])
data['currency'] = data['advertEstateOffer'].apply(lambda x: x[0]['currency'])
data['keyOfferType'] = data['advertEstateOffer'].apply(lambda x: x[0]['keyOfferType'])
data['keyEstateType'] = data['advertEstateOffer'].apply(lambda x: x[0]['keyEstateType'])
data['keyDisposition'] = data['advertEstateOffer'].apply(lambda x: x[0]['keyDisposition'])
data['surface'] = data['advertEstateOffer'].apply(lambda x: x[0]['surface'])
data['surfaceLand'] = data['advertEstateOffer'].apply(lambda x: x[0]['surfaceLand'])

data.head()
```

```python
# Dropping unnecessary columns

data.drop(['keyAdvertType', 'orderPriority', 'type', 'timeOrder', 'timezone_type', 'timezone', 'advertEstateOffer', 'keyEstateType'], axis=1, inplace=True)
data.head()
```

```python
# Exporting data

data.to_excel('bezrealitky_uri.xlsx', index=False)
```
